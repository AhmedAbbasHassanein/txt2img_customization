```python
from diffusers import DiffusionPipeline

pipe = DiffusionPipeline.from_pretrained("ahmedabaashassanein/stablediffusioncustomized")
pipe = pipe.to("cuda")  # Move the model to GPU for faster inference
```

```python
def generate_image(prompt):
    # Generate an image from the text prompt
    image = pipe(prompt).images[0]
    return image
```

```python
import gradio as gr

# Create the Gradio interface
iface = gr.Interface(
    fn=generate_image,         # Function that generates images
    inputs="text",             # Input is a text prompt
    outputs="image",           # Output is an image
    title="Text-to-Image Generator",
    description="Enter a prompt and generate an image using Stable Diffusion!"
)

# Launch the interface
iface.launch(share=True)
```
